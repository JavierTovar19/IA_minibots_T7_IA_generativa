```python
# =========================
# 1. CONECTAR GOOGLE DRIVE
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# 2. LIBRERÍAS
# =========================
import os
from PyPDF2 import PdfReader
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# =========================
# 3. LEER TODOS LOS PDFs DE UNA CARPETA
# =========================
def leer_multiples_pdfs(carpeta):
    texto_total = ""

    for archivo in os.listdir(carpeta):
        if archivo.endswith(".pdf"):
            ruta = os.path.join(carpeta, archivo)
            print(f"📄 Leyendo: {archivo}")

            reader = PdfReader(ruta)

            for pagina in reader.pages:
                contenido = pagina.extract_text()
                if contenido:
                    texto_total += contenido + "\n"

    return texto_total

# =========================
# 4. DIVIDIR TEXTO EN CHUNKS
# =========================
def dividir_texto(texto, tamaño=200):
    palabras = texto.split()
    chunks = []

    for i in range(0, len(palabras), tamaño):
        chunk = " ".join(palabras[i:i+tamaño])
        chunks.append(chunk)

    return chunks

# =========================
# 5. RUTA DE LA CARPETA EN DRIVE
# =========================
carpeta = "/content/drive/MyDrive/pdfs"

# =========================
# 6. PROCESAR PDFs
# =========================
texto_pdf = leer_multiples_pdfs(carpeta)

if not texto_pdf.strip():
    print("⚠️ No se pudo extraer texto de los PDFs.")
    exit()

documentos = dividir_texto(texto_pdf)

print(f"\n✅ Se generaron {len(documentos)} fragmentos de texto\n")

# =========================
# 7. VECTORIZACIÓN
# =========================
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documentos)

# =========================
# 8. FUNCIÓN RAG
# =========================
def buscar_respuesta(pregunta):
    pregunta_vec = vectorizer.transform([pregunta])
    similitudes = cosine_similarity(pregunta_vec, X)

    indice = similitudes.argmax()
    mejor_score = similitudes[0][indice]

    if mejor_score < 0.1:
        return "No encontré información relevante en los documentos."

    return "📚 Según los documentos: " + documentos[indice]

# =========================
# 9. INTERFAZ TIPO CHAT
# =========================
print("🤖 Chatbot Universidad (múltiples PDFs)")
print("Escribe 'salir' para terminar\n")

while True:
    pregunta = input("Tú: ")

    if pregunta.lower() == "salir":
        print("Bot: ¡Hasta luego!")
        break

    respuesta = buscar_respuesta(pregunta)
    print("Bot:", respuesta, "\n")
```
