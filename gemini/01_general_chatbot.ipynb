```python
import os
import getpass
import google.generativeai as genai
import gradio as gr

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("API KEY:")
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
model = genai.GenerativeModel('gemini-1.5-flash', system_instruction="Eres un asistente amable y conciso. Responde en español.")
chat = model.start_chat()
```

```python
def responder_general(mensaje, historial):
    try:
        response = chat.send_message(mensaje)
        return response.text
    except Exception as e:
        return f"Error: {e}"

gr.ChatInterface(responder_general, title="Chatbot General (Nativo Gemini SDK)").launch(share=True)
```
